## Cleaning Data



1.URLs 
2.Hastags
3.mentions 
4.special characters
5.punctuations

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+', ' ', txt)
    cleanText = re.sub('#\S+', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

## Vectorization 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredText=tfidf.transform(df['Resume'])
```

```python
requiredText
```

## Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(requiredText,df['Category'],
                            test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Model training 

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

knn=OneVsRestClassifier(KNeighborsClassifier())
knn.fit(x_train,y_train)

y_pred=knn.predict(x_test)
```

```python
print("yprediction ", y_pred)
print(len(y_pred))
```

```python
# the accuracy score 
print("accuracy score ", accuracy_score(y_test,y_pred))
```

## Making the prediction System 

```python
def prediction_system(resume):
    cleaned_text = cleanResume(resume) 
    vectorized_text = tfidf.transform([cleaned_text])
    vectorized_text = vectorized_text.toarray()
    predicted_category = svc_model.predict(vectorized_text)
    predicted_category_name = le.inverse_transform(predicted_category)
    return predicted_category_name[0]
```
